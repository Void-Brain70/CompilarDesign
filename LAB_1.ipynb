```python
import re # RexEx Module used for regular expressions

source_file = input("Enter filename or path-> ") # given input source file
fp = open(source_file, 'r')  # this function will open file with name “filename” in specified “mode”.
test = fp.read() # for returns the specified number of bytes from the file

oparators = ['+', '-', '*', '/', '=', '+=', '-=', '==', '<', '>', '<=', '>='] # define arithmetic oparators
keywords = ['auto','break', 'case', 'char', 'const', 'continue', 'default', 'do', 
			'double', 'else', 'enum', 'extern', 'float', 'for', 'goto', 'if', 
			'int', 'long', 'register', 'return', 'short', 'signed', 'sizeof', 'static', 
			'struct', 'switch', 'typedef', 'union', 'unsigned', 'void', 'volatile', 'while'] # define keywords
punctuationandparenthesis = [' ', '	', '.', ',', '\n', ';', '(', ')', '<', '>', '{', '}', '[', ']'] # define Parenthesis and Punctuation 
symbols = ['!', '@', '#', '$', '%', '&', '^', '*'] # define symbols


opa = [] # list define for arithmetic oparators
key = [] # list define for keywords
pun = [] # list define for punctuation and parenthesis
sym = [] # list define for Symbols
ide = [] # list define for Any word either combination of characters and digits or combination of characters: Identifier.
con = [] # list define for Any number : Constant

tokens = [] # list define for add number of tokens
Str = False # check string 
Word = False # check word
Cmt = 0 # add number of comment
token = '' # speace of token

# Check given input have any string,word,comment,space,symbol
for i in test:
	if i == '/':
		Cmt = Cmt+1

	elif Cmt == 2:
		if i == '\n':
			token = ''
			Cmt = 0
	#checking given input have any sting if have then add string in token[] list
	elif i == '"' or i == "'":
		if Str:
			tokens.append(token) 
			token = ''
		Str = not Str 

	elif Str:
		token = token+i
  #checking given input have any symbol if have then add symbol in token[] list  
	elif i in symbols:
		tokens.append(i) 
	
  #checking given input have any number or word if have then add this number and word in token[] list         
	elif i.isalnum() and not Word: # isalnum() is a built-in Python function that checks whether all characters in a string are alphanumeric.
		Word = True
		token = i
  #checking given input have any punctuation and parenthesis or at a time oparator if have then added  in token[] list  
	elif (i in punctuationandparenthesis) or (i in oparators):
		if token:
			tokens.append(token)
			token = ''
        
		if not (i==' ' or i=='\n' or i=='	'):
			tokens.append(i)

	elif Word:
		token = token+i


for token in tokens:
	if token in symbols:
		sym.append(token)

	elif token in oparators:
		opa.append(token)

	elif token in keywords:
		key.append(token)
				
	elif re.search("^[_a-zA-Z][_a-zA-Z0-9]*$",token):
		ide.append(token)
		
	elif token in punctuationandparenthesis:
		pun.append(token)
		
	else:
		con.append(token)
	
						
print("Number of tokens->", len(tokens))   

print("\nNumber of keywords-> ",len(key))
print(key);

print("\nNumber of special symbols-> ",len(sym))
print(sym);

print("\nNumber of Arithmetic oparators-> ",len(opa))
print(opa);

print("\nNumber of identifiers-> ",len(ide))
print(ide);

print("\nNumber of constants = ",len(con))
print(con);

print("\nNumber of Punctuation and Parenthesis  = ",len(pun))
print(pun);

fp.close()
```
