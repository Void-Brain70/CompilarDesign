```python
import re # RexEx Module used for regular expressions

source_file = input("Enter filename or path-> ") # given input source file
fp = open(source_file, 'r')  # this function will open file with name “filename” in specified “mode”.
text = fp.read() # for returns the specified number of bytes from the file

symbols = ['!', '@', '#', '$', '%', '&', '^', '*']
oparators = ['+', '-', '*', '/', '=', '+=', '-=', '==', '<', '>', '<=', '>=']
keywords = ['auto','break', 'case', 'char', 'const', 'continue', 'default', 'do', 
			'double', 'else', 'enum', 'extern', 'float', 'for', 'goto', 'if', 
			'int', 'long', 'register', 'return', 'short', 'signed', 'sizeof', 'static', 
			'struct', 'switch', 'typedef', 'union', 'unsigned', 'void', 'volatile', 'while']
punctuationandparenthesis = [' ', '	', '.', ',', '\n', ';', '(', ')', '<', '>', '{', '}', '[', ']']


is_keywords = []
is_spl_symbols = []
is_oparators = []
is_punctuationandparenthesis = []
is_identifiers = []
is_constants = []

tokens = []
isStr = False
isWord = False
isCmt = 0
token = ''

for i in text:
	if i == '/':
		isCmt = isCmt+1

	elif isCmt == 2:
		if i == '\n':
			token = ''
			isCmt = 0
	
	elif i == '"' or i == "'":
		if isStr:
			tokens.append(token)
			token = ''
		isStr = not isStr 

	elif isStr:
		token = token+i
    
	elif i in symbols:
		tokens.append(i)
           
	elif i.isalnum() and not isWord:
		isWord = True
		token = i
    
	elif (i in punctuationandparenthesis) or (i in oparators):
		if token:
			tokens.append(token)
			token = ''
        
		if not (i==' ' or i=='\n' or i=='	'):
			tokens.append(i)

	elif isWord:
		token = token+i


for token in tokens:
	if token in symbols:
		is_spl_symbols.append(token)

	elif token in oparators:
		is_oparators.append(token)

	elif token in keywords:
		is_keywords.append(token)
				
	elif re.search("^[_a-zA-Z][_a-zA-Z0-9]*$",token):
		is_identifiers.append(token)
		
	elif token in punctuationandparenthesis:
		is_punctuationandparenthesis.append(token)
		
	else:
		is_constants.append(token)
	
						
print("Number of tokens->", len(tokens))   
print("\nNumber of keywords-> ",len(is_keywords))
print(is_keywords);
print("\nNumber of special symbols-> ",len(is_spl_symbols))
print(is_spl_symbols);
print("\nNumber of Arithmetic oparators-> ",len(is_oparators))
print(is_oparators);
print("\nNumber of identifiers-> ",len(is_identifiers))
print(is_identifiers);
print("\nNumber of constants = ",len(is_constants))
print(is_constants);
print("\nNumber of Punctuation and Parenthesis  = ",len(is_punctuationandparenthesis))
print(is_punctuationandparenthesis);
fp.close()
```
